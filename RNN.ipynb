```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
# import re
# def special_characters(text):
#   return re.sub(r'[^a-zA-Z0-9\s]','',text)
# data= special_characters(data)
```

```python
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
input_data=[]
for i in data.split('\n'):
  integer_sequences=tokenizer.texts_to_sequences([i])[0]  # Converting text to integer sequences

  for n in range(1,len(integer_sequences)):
    input_data.append(integer_sequences[:n+1])
```

```python
input_data
```

```python
max_len=max([len(x) for x in input_data])
```

```python
max_len
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
paded_inputs=pad_sequences(input_data,maxlen=max_len,padding='pre')
```

```python
paded_inputs
```

```python
X=paded_inputs[:,:-1]
y=paded_inputs[:,-1]
```

```python
from keras.utils import to_categorical
```

```python
len(tokenizer.word_index)
```

```python
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
print(X.shape)
print(y.shape)
```

```python
X
```

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense
```

```python
print(len(tokenizer.word_index)+1)
```

```python
model=Sequential()
model.add(Embedding(255,100))
model.add(LSTM(128))
model.add(Dense(255,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.weights  # This will return the list of weight arrays

for weight in model.weights:
    print(weight)
```

```python
model.fit(X,y,epochs=100)
```
